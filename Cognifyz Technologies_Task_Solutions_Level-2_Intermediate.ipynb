```python
#To view the 5 sample data
df.sample(5)
```

```python
#To view the Investment Avenues data
df['Avenue']
```

```python
#to view the info of the Avenue column
df["Avenue"].info()
```

```python
#To check any null values
df["Avenue"].isnull().sum()
```

```python
#To view the summary of Avenue column
df["Avenue"].describe()
```

```python
#To view the unique values in the Avenue column
df["Avenue"].unique()
```

```python
#To view the indivitual values counts in the column
df["Avenue"].value_counts()
```

```python
#To plot the countplot for the graphical representation of the column
cx=sns.countplot(x="Avenue", data=df)
for i in cx.containers:
    cx.bar_label(i,)
plt.title("Distribution of Avenue")
plt.show()
```

```python
#To plot pie chart
gender_distribution= df['Avenue'].value_counts()
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Avenue')
plt.show()
```

Analysing Summary:

      a)There are 40 rows in the column
      
      b)There are 4 unique values in the column
               ->Mutual Fund
               ->Equity
               ->Fixed Deposits
               ->Public Provident Fund
      c)The most common repeated value is Mutual Fund
      
3)From the value counts it shows that the Mutual Fund has the highest number of count and Public Provident Fund has the lowest number of count

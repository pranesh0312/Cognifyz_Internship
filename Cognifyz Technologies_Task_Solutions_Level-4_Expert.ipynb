# Task 8: Investment Duration

1)Analyze Investment Duration Column: Examine the column containing information about participants' investment durations.

2)Calculate Average Duration: Use appropriate statistical methods to calculate the average investment duration.

```python
#To view the 5 sample data
df.sample(5)
```

```python
#To view the Duration column
df["Duration"]
```

```python
#to view the info of the Duration column
df["Duration"].info()
```

```python
#To check any null values
df["Duration"].isnull().sum()
```

```python
#To view the summary of Duration column
df["Duration"].describe()
```

```python
#To view the unique values in the column
df["Duration"].unique()
```

```python
#To view the indivitual values counts in the column
df["Duration"].value_counts()
```

```python
#To view the mode(frequent repeated value) in the column
df["Duration"].mode()
```

```python
#Converting to numerical values by using replace function
df["Duration"][df["Duration"]=="1-3 years"]=df["Duration"][df["Duration"]=="1-3 years"].replace('1-3 years',2).astype(float)
```

```python
df["Duration"][df["Duration"]=="3-5 years"]=df["Duration"][df["Duration"]=="3-5 years"].replace('3-5 years',4).astype(float)
```

```python
df["Duration"][df["Duration"]=="Less than 1 year"]=df["Duration"][df["Duration"]=="Less than 1 year"].replace('Less than 1 year',0.5).astype(float)
```

```python
df["Duration"][df["Duration"]=="More than 5 years"]=df["Duration"][df["Duration"]=="More than 5 years"].replace('More than 5 years',8).astype(float)
```

```python
#Duration column after converting to numerical values
df["Duration"]=df["Duration"].astype(float)
df["Duration"].head(5)
```

```python
#To view the Summary of Duration column
df["Duration"].describe()
```

```python
#To view the average Duration
df["Duration"].mean()
```

```python
#to plot Histogram to view the distribution of duration column
a=sns.histplot(df["Duration"])
for i in a.containers:
    a.bar_label(i,)
plt.title("Distribution of Duration")
plt.show()
```

```python
xy=sns.countplot(data=df,x=df["Duration"])
for i in xy.containers:
    xy.bar_label(i,)
plt.title("Duration")
plt.show()
```

From the analysing,it shows that 1-3 and 3-5 Years duration is the most common Duration participants rely on
